```python
import pandas as pd
from sklearn.datasets import load_iris
from scipy import stats
iris = load_iris()
iris_ds = pd.DataFrame(data=iris.data, columns=iris.feature_names)
sepalLengths = iris_ds['sepal length (cm)']

# Calculate mean, median, and mode
mean_sepalLength = sepalLengths.mean()
median_sepalLength = sepalLengths.median()
mode_sepalLength = stats.mode(sepalLengths)

# Access the mode value directly
mean_sepalLength, median_sepalLength, mode_sepalLength.mode
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
sepal_lengths = iris_df['sepal length (cm)']
bins = np.linspace(sepal_lengths.min(), sepal_lengths.max(), 10)

# Calculate the histogram
hist, bin_edges = np.histogram(sepal_lengths, bins=bins, density=True)
probability_distribution = hist / hist.sum()
probability_distribution_df = pd.DataFrame({
    'sepal_length_bin': bin_edges[:-1],
    'probability': probability_distribution
})
print(probability_distribution_df)

# Plot the probability distribution
plt.figure(figsize=(10, 6))
plt.bar(bin_edges[:-1], probability_distribution, width=np.diff(bin_edges), edgecolor='black', align='edge')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Probability')
plt.title('Probability Distribution of Sepal Lengths')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Extract the petal lengths
petal_lengths = iris_df['petal length (cm)']

# Sort the petal lengths
sorted_petal_lengths = np.sort(petal_lengths)

# Calculate the CDF values
cdf_values = np.arange(1, len(sorted_petal_lengths) + 1) / len(sorted_petal_lengths)

# Create a DataFrame for the CDF
cdf_df = pd.DataFrame({
    'petal_length': sorted_petal_lengths,
    'cdf': cdf_values
})

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(cdf_df['petal_length'], cdf_df['cdf'], marker='.', linestyle='none')
plt.xlabel('Petal Length (cm)')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset('iris')
sepal_widths = iris['sepal_width']
sns.distplot(sepal_widths, hist=False, kde=True,
             bins=int(len(sepal_widths) / 10), color='darkblue',
             hist_kws={'edgecolor': 'black'},
             kde_kws={'linewidth': 2})

plt.title('Probability Density Function (PDF) of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris')

# Assuming you want to find the probability of petal length being greater than 4.0
given_value = 4.0

petal_lengths = iris['petal_length']

prob_greater_than_given = sum(petal_lengths > given_value) / len(petal_lengths)

print(f"Probability of petal length > {given_value}: {prob_greater_than_given}")
```

```python
from scipy import stats
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
petal_length_setosa = iris.loc[iris['species'] == 'setosa', 'petal_length']
petal_length_versicolor = iris.loc[iris['species'] == 'versicolor', 'petal_length']
t_statistic, p_value = stats.ttest_ind(petal_length_setosa, petal_length_versicolor)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
sepal_length = iris['sepal_length']
sepal_width = iris['sepal_width']
covariance = sepal_length.cov(sepal_width)
print("Covariance between sepal length and sepal width:", covariance)
correlation = sepal_length.corr(sepal_width)
print("Correlation between sepal length and sepal width:", correlation)
```
