```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/titanic.csv.csv')
categories = data.iloc[:,0].astype(str)
for column in data.columns[1:]:
  frequencies = pd.to_numeric(data[column], errors='coerce')
  plt.figure(figsize=(10,5))
  plt.bar(categories, frequencies)
  plt.xlabel('Categories')
  plt.ylabel('Frequencies')
  plt.title(column)
  plt.grid(axis='y')
  plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools

# Load the dataset
data = pd.read_csv('/content/titanic.csv.csv')

# Get the list of columns
columns = data.columns

# Iterate through each pair of columns
for col1, col2 in itertools.combinations(columns, 2):
    x = data[col1]
    y = data[col2]

    # Create a scatter plot
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='blue', marker='o')

    # Add titles and labels
    plt.title(f'Scatter Plot of {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)

    # Show the plot
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools

data = pd.read_csv('/content/titanic.csv.csv')

columns = data.columns

for col1, col2 in itertools.combinations(columns, 2):
    x = data[col1]
    y = data[col2]
    if pd.api.types.is_numeric_dtype(x) and pd.api.types.is_numeric_dtype(y):
        plt.figure(figsize=(10, 5))
        plt.scatter(x, y, color='blue', marker='o')

        plt.title('Scatter Plot of {col1} vs {col2}')
        plt.xlabel(col1)
        plt.ylabel(col2)

        plt.grid(True)
        plt.show()
    else:
        print("Skipping scatter plot for {col1} vs {col2} due to non-numeric data.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

# Select only numeric columns for correlation calculation
numeric_tips = tips.select_dtypes(include=['number'])
corr_matrix = numeric_tips.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].hist(tips['total_bill'], bins=20, color='blue', edgecolor='black')
axs[0, 0].set_title('Histogram of Total Bill')
axs[0, 0].set_xlabel('Total Bill')
axs[0, 0].set_ylabel('Frequency')
axs[0, 1].scatter(tips['total_bill'], tips['tip'], color='green', edgecolor='black')
axs[0, 1].set_title('Scatter Plot of Total Bill vs Tip')
axs[0, 1].set_xlabel('Total Bill')
axs[0, 1].set_ylabel('Tip')
sns.boxplot(x='day', y='total_bill', data=tips, ax=axs[1, 0])
axs[1, 0].set_title('Box Plot of Total Bill by Day')
axs[1, 0].set_xlabel('Day')
axs[1, 0].set_ylabel('Total Bill')
sns.barplot(x='day', y='total_bill', data=tips, ax=axs[1, 1], ci=None, palette='viridis')
axs[1, 1].set_title('Bar Plot of Average Total Bill by Day')
axs[1, 1].set_xlabel('Day')
axs[1, 1].set_ylabel('Average Total Bill')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
custom_palette = ['#FF6347', '#1E90FF', '#32CD32', '#FFD700']
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips, palette=custom_palette)
plt.title('Average Total Bill by Day', fontsize=16)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Average Total Bill', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=tips, palette='muted')
plt.title('Distribution of Total Bill by Day', fontsize=16)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Total Bill', fontsize=14)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Total Bill with KDE Overlay', fontsize=16)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```
