```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
# S1: Here we first load the iris dataset
iris = load_iris()
# S2: Prepare the data
u = iris.data  # Features
v = iris.target  # indicating target variables
# Spliting the data into training and testing sets (80:20 ratio)
u_train, u_test, v_train, v_test = train_test_split(u, v, test_size=0.2, random_state=42)
# S3: Trainning the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(u_train, v_train)
# S4: Evaluate the model
v_pred = clf.predict(u_test)
print("Classification Report:")
print(classification_report(v_test, v_pred))
print("Confusion Matrix:")
print(confusion_matrix(v_test, v_pred))
# S5: Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree for Iris Dataset")
plt.show()
```
